```python
import csv
import json
import pandas as pd
import requests
```

```python
queries = ['funko', 'figure', 'boneco']
ids = []
limit = 50
for query in queries:
  url = f"https://api.mercadolibre.com/sites/MLA/search?q={query}&limit={limit}#json"
  response = requests.get(url)
  data = response.json()
  for item in data['results']:
    ids.append(item['id'])

```

```python
len(ids)
```

```python
items = []
for id in ids:
  url = f"https://api.mercadolibre.com/items/{id}"
  response = requests.get(url)
  data = response.json()
  items.append(data)
```

```python
items[0]
clear_items = []
for item in items:
  id = item['id']
  name = item['title']
  category_id = item['category_id']
  price = item['price']
  clear_item = {
      'id': id,
      'name': name,
      'category_id': category_id,
      'price': price
  }
  clear_items.append(clear_item)
```

```python
df = pd.DataFrame(clear_items)
df.head()
```

```python
df.to_csv('output_file.csv', index=False)
```
